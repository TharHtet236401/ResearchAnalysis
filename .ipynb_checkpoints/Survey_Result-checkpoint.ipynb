```python
columns_to_remove = [
    'Timestamp', 
    'Have you ever taken the treatment of hospitals or clinics?',
    'Do you think that your heathcare data are being used for research?',
    'Enter Your Name  ( Optional) '
]

```

```python
data_cleaned = data.drop(columns=columns_to_remove)
```

```python
data_cleaned = data_cleaned.dropna()
```

```python
data_cleaned.to_csv('data/answers_cleaned.csv', index=False)
```

```python
data_cleaned.describe(include = "all")
```

```python
frequency_counts = data_cleaned["Do you agree that the patient data being used WITHOUT Informed consent is acceptable even for good purposes (like in COVID - 19) "].value_counts()
print(frequency_counts)
```

```python
import matplotlib.pyplot as plt

max_count = frequency_counts.max()
max_index = frequency_counts.idxmax()

colors = ['skyblue' if index != max_index else 'red' for index in frequency_counts.index]

plt.figure(figsize=(10, 6))
bars = plt.bar(frequency_counts.index, frequency_counts, color=colors)
plt.title('Importance of Transparency Level')
plt.xlabel('Awarenees Level')
plt.ylabel('Participants')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding text label for the highest bar
for bar in bars:
    yval = bar.get_height()
    if bar.get_height() == max_count:
        plt.text(bar.get_x() + bar.get_width()/2, yval + 1, int(yval), ha='center', va='bottom', color='red')

plt.show()
```

```python
plt.figure(figsize=(6, 6))
# Create the pie chart with empty labels
plt.pie(frequency_counts, labels=frequency_counts.index, autopct='%1.1f%%', startangle=140)
# Add a legend

plt.title('Importanance of Consent for Good Purposes')
plt.show()
```
